```python
from langchain_community.embeddings import OllamaEmbeddings

```

```python
embeddings = OllamaEmbeddings(
    model="gemma:2b"
    )
    
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
         "This is a document about banana",
         "This is a document about orange",
        
        ]
    )
```

```python
print(r1[0])
```

```python
print(r1[1])
```
